## Gender and Age Distributions among Titanic Passengers
### Dataset Description:
The Titanic dataset is a commonly used dataset in data science and machine learning projects. It contains information about the passengers on the Titanic, including their demographics, ticket details, and survival status. 
### The dataset columns are described below:
##### PassengerId: A unique identifier for each passenger.
##### Survived: Indicates whether the passenger survived (1) or not (0).
##### Pclass: Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd).
##### Name: The full name of the passenger.
##### Sex: Gender of the passenger (male/female).
##### Age: Age of the passenger in years. Some entries contain NaN values where the age is not available.
##### SibSp: Number of siblings or spouses aboard the Titanic.
##### Parch: Number of parents or children aboard the Titanic.
##### Ticket: Ticket number of the passenger.
##### Fare: The fare paid by the passenger for the ticket.
##### Cabin: Cabin number where the passenger stayed. This column has many missing values.
##### Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

## Visualizations
#### Gender Distribution on Titanic:
A bar chart illustrating the distribution of passengers based on gender. The plot shows the count of male and female passengers, providing a clear view of gender demographics on the Titanic.
#### Age Distribution on Titanic:
A histogram displaying the distribution of passengers' ages. This visualization helps in understanding the age demographics, showing the frequency of different age groups among the passengers.

These visualizations help to get an initial understanding of the dataset and are often the first step in exploratory data analysis (EDA). 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = pd.read_csv('Titanic-Dataset.csv')

# Bar Chart for Gender Distribution
gender_counts = df['Sex'].value_counts()
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Gender Distribution on Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Histogram for Age Distribution
plt.figure(figsize=(8, 6))
df['Age'].plot(kind='hist', bins=30, color='c', edgecolor='black')
plt.title('Age Distribution on Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
